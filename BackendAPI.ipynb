### Molecule / Project API

#### Create Project

```python
# Header per autenticazione e tipo di contenuto
headers = {
    "Authorization": f"Bearer {access_token}",  # oppure "Token {token}" se usi TokenAuth
    "Content-Type": "application/json",
}

# Dati del nuovo progetto
data = {
    "name": "My Project"
}

# Invio della richiesta POST
response = requests.post("http://127.0.0.1:8000/api/projects/", json=data, headers=headers)

# Visualizzo il risultato
print("Status:", response.status_code)
try:
    print("Response JSON:", response.json())
except:
    print(response.text)
```

#### Edit Project name

```python
project_id = 9
url = "http://127.0.0.1:8000/api/projects/"
#url = "https://eleanore-arkosic-regrettingly.ngrok-free.dev/api/projects/"

headers = {
    "Authorization": f"Bearer {access_token}",  # oppure "Token {token}" se usi TokenAuth
    "Content-Type": "application/json",
}

# Dati del nuovo progetto
data = {
    "name": "My Project renamed"
}

# Invio della richiesta POST
response = requests.patch(f"{url}{project_id}/", json=data, headers=headers)

# Visualizzo il risultato
print("Status:", response.status_code)
try:
    print("Response JSON:", response.json())
except:
    print(response.text)
```

### Access API

#### Register

```python
# Register - Esempio di registrazione
import os

url = "http://127.0.0.1:8000/api/auth/register/"

# Usa variabili di test, non credenziali reali
test_email = os.getenv("TEST_EMAIL", "test-user@molbook.local")
test_password = os.getenv("TEST_PASSWORD", "TestPassword123!")

payload = {
    "email": test_email,
    "password": test_password,
    "first_name": "Test",
    "last_name": "User",
    "date_of_birth": "1990-01-01",
    "gender": "M"
}

resp = requests.post(url, json=payload, headers={"Accept-Language": "en-us"})
print(resp.status_code)
print(resp.json())
```

#### Login

```python
# Login - Usa environment variables
import os

url = "http://127.0.0.1:8000/api/auth/login/"

# Leggi credenziali da environment variables
username = os.getenv("API_TEST_USERNAME", "demo@molbook.it")
password = os.getenv("API_TEST_PASSWORD", "django")

payload = {
    "email": username,   
    "password": password,
}

resp_login = requests.post(url, json=payload)
print("Status:", resp_login.status_code)
print("JSON:", resp_login.json())

if resp_login.status_code == 200:
    access_token = resp_login.json()["access"]
    print("✅ Access token acquired successfully")
```
